```python
%pylab inline
import pommerman
import random

random.seed(0)
config='PommeFFACompetition-v0'
_agent_list = [pommerman.agents.SimpleAgent() for _ in range(4)]
env = pommerman.make(config, _agent_list)
obs = env.reset()
for _ in range(3):
    actions = env.act(obs)
    obs, reward, done, info = env.step(actions)
```
