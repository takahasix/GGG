```python
%pylab inline
import pommerman
from pommerman import constants
config='PommeFFACompetition-v0'
```

```python
_agent_list = [pommerman.agents.SimpleAgent() for _ in range(4)]
env = pommerman.make(config, _agent_list)
env._num_rigid = 0
env._num_items = 0
env._num_wood = 0
for agent in env._agents:
    agent.can_kick = True
obs = env.reset()
env.render()
actions = [constants.Action.Right.value,
           constants.Action.Right.value,
           constants.Action.Left.value,
           constants.Action.Left.value]
env.step(actions)
env.render()
actions = [constants.Action.Bomb.value,
           constants.Action.Up.value,
           constants.Action.Up.value,
           constants.Action.Left.value]
env.step(actions)
env.render()
actions = [constants.Action.Stop.value,
           constants.Action.Bomb.value,
           constants.Action.Stop.value,
           constants.Action.Left.value]
env.step(actions)
env.render()
actions = [constants.Action.Up.value,
           constants.Action.Down.value,
           constants.Action.Stop.value,
           constants.Action.Down.value]
env.step(actions)
env.render()
actions = [constants.Action.Down.value,
           constants.Action.Stop.value,
           constants.Action.Stop.value,
           constants.Action.Down.value]
env.step(actions)
env.render()
actions = [constants.Action.Stop.value,
           constants.Action.Up.value,
           constants.Action.Stop.value,
           constants.Action.Down.value]
env.step(actions)
env.render()
actions = [constants.Action.Stop.value,
           constants.Action.Stop.value,
           constants.Action.Stop.value,
           constants.Action.Bomb.value]
env.step(actions)
env.render()
actions = [constants.Action.Stop.value,
           constants.Action.Stop.value,
           constants.Action.Stop.value,
           constants.Action.Up.value]
env.step(actions)
env.render()
actions = [constants.Action.Stop.value,
           constants.Action.Stop.value,
           constants.Action.Stop.value,
           constants.Action.Down.value]
env.step(actions)
env.render()
for _ in range(2):
    actions = [constants.Action.Stop.value] * 4
    env.step(actions)
    env.render()
actions = [constants.Action.Stop.value,
           constants.Action.Stop.value,
           constants.Action.Left.value,
           constants.Action.Stop.value]
env.step(actions)
env.render()
for _ in range(10):
    actions = [constants.Action.Stop.value] * 4
    env.step(actions)
    env.render()
```

```python
env.close()
```
